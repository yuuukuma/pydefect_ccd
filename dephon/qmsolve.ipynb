```python
from qmsolve import eV, Hamiltonian, SingleParticle
from qmsolve import Å


#define the interaction potential
def harmonic_oscillator(particle):
    k = 100. * eV / Å**2
    return 0.5 * k * particle.x**2

particle = SingleParticle(m=1.0)

#define the Hamiltonian
H = Hamiltonian(particles=particle,
                potential=harmonic_oscillator,
                spatial_ndim=1, N=512, extent=20*Å)
```

```python

eigenstates = H.solve(max_states = 30)
print(eigenstates.energies) # The printed energies are in eV.
```

```python
from qmsolve import init_visualization

# Visualize the Eigenstates
visualization = init_visualization(eigenstates)
visualization.slider_plot() #interactive slider

# (Optional: Visualize a specific eigenstate)
# visualization.plot_eigenstate(0)
```
